```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import glob
import re

files = glob.glob('data/indexes/Clean dfs/*.csv')

index_names = []
index_list = []
for file in files:
    name = re.findall('\\\(.+).csv', file)
    name_string = ''.join(name)
    index_names.append(name_string)


for i in range(len(index_names)):
    index_names[i] = pd.read_csv('data/indexes/Clean dfs/'+index_names[i]+'.csv', index_col=None, header=0)
    
```

```python
cac40 = index_names[0]
chinaa50 = index_names[1]
dax = index_names[2]
dowjones = index_names[3]
eurostoxx = index_names[4]
ibex = index_names[5]
nasdaq = index_names[6]
```

```python
#cac40 - averages of variation

var_averages = []
var_averages3 = []

for row in range(len(cac40['Var_%'])):
            
    avg = np.mean(cac40['Var_%'][row+1:row+6])
    var_averages.append(round(avg,3))
    
    avg3 = np.mean(cac40['Var_%'][row+1:row+4])
    var_averages3.append(round(avg3,3))
    
    
cac40['cac40_last_5_avg'] = var_averages
cac40['cac40_last_3_avg'] = var_averages3
cac40['cac40_last_1'] = cac40['Var_%'].shift(-1)
```

```python
#cac40 - averages of last index

last_averages = []
last_averages3 = []

for row in range(len(cac40['Last'])):
            
    avg = np.mean(cac40['Last'][row+1:row+6])
    last_averages.append(round(avg,3))
    
    avg3 = np.mean(cac40['Last'][row+1:row+4])
    last_averages3.append(round(avg3,3))
    
    
cac40['cac40index_last_5'] = last_averages
cac40['cac40index_last_3'] = last_averages3
cac40['cac40index_last_1'] = cac40['Last'].shift(-1)
```

## Appending indexes to companies dataframe

```python
ibex_base = ibex[['Date', 'Last']]

cac40_reduced = cac40[['Date', 'cac40_last_5_avg', 'cac40_last_3_avg', 'cac40_last_1']]
merged_dfs = ibex_base.merge(cac40_reduced, on='Date')

chinaa50_reduced = chinaa50[['Date', 'chinaa50_last_5_avg', 'chinaa50_last_3_avg', 'chinaa50_last_1']]
merged_dfs = merged_dfs.merge(chinaa50_reduced, on='Date')

dax_reduced = dax[['Date', 'dax_last_5_avg', 'dax_last_3_avg', 'dax_last_1']]
merged_dfs = merged_dfs.merge(dax_reduced, on='Date')

dowjones_reduced = dowjones[['Date', 'dowjones_last_5_avg', 'dowjones_last_3_avg', 'dowjones_last_1']]
merged_dfs = merged_dfs.merge(dowjones_reduced, on='Date')

eurostoxx_reduced = eurostoxx[['Date', 'eurostoxx_last_5_avg','eurostoxx_last_3_avg', 'eurostoxx_last_1']]
merged_dfs = merged_dfs.merge(eurostoxx_reduced, on='Date')

ibex_reduced = ibex[['Date', 'ibex_last_5_avg', 'ibex_last_3_avg', 'ibex_last_1']]
merged_dfs = merged_dfs.merge(ibex_reduced, on='Date')

nasdaq_reduced = nasdaq[['Date', 'nasdaq_last_5_avg', 'nasdaq_last_3_avg', 'nasdaq_last_1']]
merged_dfs = merged_dfs.merge(nasdaq_reduced, on='Date')

merged_dfs.head()
```

```python
correlation = merged_dfs.corr()

plt.figure(figsize=(20,20))
ax=sns.heatmap(correlation, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
