## Appending indexes to companies dataframe

```python
ibex35_reduced = ibex35[['Date','Var_%']]

cac40_reduced = cac40[['Date', 'cac40_last_5_avg', 'cac40_last_3_avg', 'cac40_last_1']]
merged_dfs = ibex35_reduced.merge(cac40_reduced, on='Date')

chinaa50_reduced = chinaa50[['Date', 'chinaa50_last_5_avg', 'chinaa50_last_3_avg', 'chinaa50_last_1']]
merged_dfs = merged_dfs.merge(chinaa50_reduced, on='Date')

dax_reduced = dax[['Date', 'dax_last_5_avg', 'dax_last_3_avg', 'dax_last_1']]
merged_dfs = merged_dfs.merge(dax_reduced, on='Date')

dowjones_reduced = dowjones[['Date', 'dowjones_last_5_avg', 'dowjones_last_3_avg', 'dowjones_last_1']]
merged_dfs = merged_dfs.merge(dowjones_reduced, on='Date')

eurostoxx_reduced = eurostoxx[['Date', 'eurostoxx_last_5_avg','eurostoxx_last_3_avg', 'eurostoxx_last_1']]
merged_dfs = merged_dfs.merge(eurostoxx_reduced, on='Date')

ibex_reduced = ibex[['Date', 'ibex_last_5_avg','ibex_last_3_avg', 'ibex_last_1']]
merged_dfs = merged_dfs.merge(ibex_reduced, on='Date')

nasdaq_reduced = nasdaq[['Date', 'nasdaq_last_5_avg', 'nasdaq_last_3_avg', 'nasdaq_last_1']]
merged_dfs = merged_dfs.merge(nasdaq_reduced, on='Date')

merged_dfs.head()
```

```python
# PCA

from sklearn.decomposition import PCA

numericData = merged_nonulls.drop(columns=['Date', 'Var_%'])

pca = PCA(n_components=7) 
principalComponents = pca.fit_transform(np.array(numericData))

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC 1', 'PC 2', 'PC 3', 'PC 4', 'PC 5', 'PC 6', 'PC 7'])


principalDf.head() 
```

```python
#Calculating percentage of variance explained by the 4 components

sum(pca.explained_variance_ratio_)
```

```python
red_merged = merged_nonulls[['Date', 'Var_%']]
red_merged['PC 1'] = principalDf['PC 1']
red_merged['PC 2'] = principalDf['PC 2']
red_merged['PC 3'] = principalDf['PC 3']
red_merged['PC 4'] = principalDf['PC 4']
red_merged['PC 5'] = principalDf['PC 5']
red_merged['PC 6'] = principalDf['PC 6']
red_merged['PC 7'] = principalDf['PC 7']
```

```python
principalDf['y'] = merged_nonulls['Var_%']
principalDf.head()
```

```python
from sklearn.model_selection import train_test_split

X = principalDf.drop(columns='y')
y = principalDf['y']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

linreg.fit(X_train, y_train)

print(linreg.intercept_)
print(linreg.coef_)

```

```python
y_pred = linreg.predict(X_test)
```

```python
df_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_pred.head()
```

```python
plt.scatter(y_test, y_pred,  color='gray')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
```
