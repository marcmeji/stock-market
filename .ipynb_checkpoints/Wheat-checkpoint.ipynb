```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
wheat = pd.read_csv('data/wheat.csv')
wheat.head()
```

```
wheat.columns = ['Date', 'Last', 'Open', 'Max', 'Min', 'Vol', 'Var_%']
```

```
wheat.shape
```

```
wheat['Date'] = pd.to_datetime(wheat['Date'], format='%d.%m.%Y')
wheat.head()
```

```
def number_formatting(x):
    a = x.replace('.','')
    return a.replace(',','.')

for col in ['Last', 'Open', 'Max', 'Min']:
    wheat[col] = wheat[col].apply(number_formatting)
    wheat[col] = pd.to_numeric(wheat[col])
    
wheat.head()
```

```
wheat.Vol.unique()
```

```
def volume_change(x):
    if 'M' in x:
        a = x.replace('K','')
        return a.replace(',','.')
    
            
wheat['Vol'] = pd.to_numeric(wheat['Vol'].apply(volume_change))
```

```
wheat.head()
```

```
def var_reformatting(x):
    a = x.replace(',','.')
    return a.replace('%','')

wheat['Var_%'] =pd.to_numeric(wheat['Var_%'].apply(var_reformatting))
```
