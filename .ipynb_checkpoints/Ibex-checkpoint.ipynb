```python
import pandas as pd
import numpy as np
```

```python
ibex = pd.read_csv("C:/Users/mmejias/Desktop/Projects Data/Project IBEX35/IBEX_10years.csv")
```

```python
ibex.head()
```

```python
ibex.drop(0, axis=0, inplace=True)
```

```python
ibex.head()
```

```python
ibex.columns = ['Date', 'Last', 'Open', 'Max', 'Min', 'Vol_M', 'Var_%']
```

```python
ibex.shape
```

```python
ibex.info()
```

```python
ibex['Date'] = pd.to_datetime(ibex['Date'], format='%d.%m.%Y')
ibex.head()
```

```python
def number_formatting(x):
    a = x.replace('.','')
    return a.replace(',','.')

for col in ['Last', 'Open', 'Max', 'Min']:
    ibex[col] = ibex[col].apply(number_formatting)
    ibex[col] = pd.to_numeric(ibex[col])
    
ibex.head()
```

```python
ibex['Vol_M'].value_counts().head()
```

```python
def volume_change(x):
    if 'M' in x:
        a = x.replace('M','')
        return a.replace(',','.')
    if 'B' in x:
        a = x.replace('B','')
        return a.replace(',','000.')
            
ibex['Vol_M'] = pd.to_numeric(ibex['Vol_M'].apply(volume_change))
```

```python
ibex.head()
```

```python
def var_reformatting(x):
    a = x.replace(',','.')
    return a.replace('%','')

ibex['Var_%'] =pd.to_numeric(ibex['Var_%'].apply(var_reformatting))
```

```python
ibex = ibex.dropna()
```
