```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
petrol = pd.read_csv('data/petrol_brent.csv')
petrol.head()
```

```python
petrol.columns = ['Date', 'Last', 'Open', 'Max', 'Min', 'Vol', 'Var_%']
```

```python
petrol.shape
```

```python
petrol['Date'] = pd.to_datetime(petrol['Date'], format='%d.%m.%Y')
petrol.head()
```

```python
def number_formatting(x):
    a = x.replace('.','')
    return a.replace(',','.')

for col in ['Last', 'Open', 'Max', 'Min']:
    petrol[col] = petrol[col].apply(number_formatting)
    petrol[col] = pd.to_numeric(petrol[col])
    
petrol.head()
```

```python
petrol.Vol.unique()
```

```python
petrol.loc[~petrol['Vol'].str.contains('K')]
```

```python
def volume_change(x):
    if 'K' in x:
        a = x.replace('K','')
        return a.replace(',','.')
    
            
petrol['Vol'] = pd.to_numeric(petrol['Vol'].apply(volume_change))
```

```python
petrol = petrol.rename(columns={'Vol':'Vol_K'})
```

```python
petrol.head()
```

```python
def var_reformatting(x):
    a = x.replace(',','.')
    return a.replace('%','')

petrol['Var_%'] =pd.to_numeric(petrol['Var_%'].apply(var_reformatting))
```
