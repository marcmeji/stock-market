```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
soy = pd.read_csv('data/soy.csv')
soy.head()
```

```python
soy.columns = ['Date', 'Last', 'Open', 'Max', 'Min', 'Vol', 'Var_%']
```

```python
soy.shape
```

```python
soy['Date'] = pd.to_datetime(soy['Date'], format='%d.%m.%Y')
soy.head()
```

```python
def number_formatting(x):
    a = x.replace('.','')
    return a.replace(',','.')

for col in ['Last', 'Open', 'Max', 'Min']:
    soy[col] = soy[col].apply(number_formatting)
    soy[col] = pd.to_numeric(soy[col])
    
soy.head()
```

```python
soy.Vol.unique()
```

```python
def volume_change(x):
    if 'K' in x:
        a = x.replace('K','')
        return a.replace(',','.')
    
            
soy['Vol'] = pd.to_numeric(soy['Vol'].apply(volume_change))
```

```python
soy = soy.rename(columns={'Vol':'Vol_K'})
```

```python
soy.head()
```

```python
def var_reformatting(x):
    a = x.replace(',','.')
    return a.replace('%','')

soy['Var_%'] =pd.to_numeric(soy['Var_%'].apply(var_reformatting))
```
