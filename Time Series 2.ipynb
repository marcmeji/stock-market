```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path = 'data/indexes/Clean dfs/ibex.csv'
data = pd.read_csv(path, index_col=['Date'], parse_dates=['Date'])
data.head(10)
```

```python
timeseries = data[['Open', 'Max', 'Min', 'Last']]
timeseries.head()
```

```python
plt.figure(figsize=(15,8))
plt.title('Ibex35 Closing Values')
plt.xlabel('Date')
plt.ylabel('Closing value')
plt.plot(timeseries['Last'])
plt.show()
```

```python
# Time series is not stationary
# Creating a function to apply moving average model to smooth time series

def moving_avg(ts,window):
    
    rolmean = ts.rolling(window=window).mean()
    
    plt.figure(figsize=(15,8))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolmean, label='Rolling mean trend')
    
    plt.plot(ts[window:], label='Actual values')
    plt.legend(loc='best')
    plt.grid(True)
```

```python
#Smooth by the previous 5 days (by week)
moving_avg(timeseries['Last'], 5)

#Smooth by the previous month (30 days)
moving_avg(timeseries['Last'], 30)

#Smooth by previous quarter (90 days)
moving_avg(timeseries['Last'], 90)
```
