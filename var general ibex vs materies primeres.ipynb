```python
import glob
import re

files = glob.glob('data/raw materials/Clean dfs/*.csv')

index_names = []

for file in files:
    name = re.findall('\\\(.+).csv', file)
    name_string = ''.join(name)
    index_names.append(name_string)


for i in range(len(index_names)):
    index_names[i] = pd.read_csv('data/raw materials/Clean dfs/'+index_names[i]+'.csv', index_col=None, header=0)
    
```

```python
files
```

```python
copper = index_names[0]
cotton = index_names[1]
gasnat = index_names[2]
gold = index_names[3]
petrol = index_names[4]
rice = index_names[5]
soy = index_names[6]
wheat = index_names[7]
```

```python
#copper

var_averages = []
var_averages3 = []

for row in range(len(copper['Var_%'])):
            
    avg = np.mean(copper['Var_%'][row+1:row+6])
    var_averages.append(round(avg,3))
    
    avg3 = np.mean(copper['Var_%'][row+1:row+4])
    var_averages3.append(round(avg3,3))
    
    
copper['copper_last_5_avg'] = var_averages
copper['copper_last_3_avg'] = var_averages3
copper['copper_last_1'] = copper['Var_%'].shift(-1)
copper.head()
```

## Appending raw materials to companies dataframe

```python
ibex35_reduced = ibex35[['Date', 'Var_%']]

copper_reduced = copper[['Date', 'copper_last_5_avg', 'copper_last_3_avg', 'copper_last_1']]
merged_dfs = ibex35_reduced.merge(copper_reduced, on='Date')

cotton_reduced = cotton[['Date', 'cotton_last_5_avg', 'cotton_last_3_avg', 'cotton_last_1']]
merged_dfs = merged_dfs.merge(cotton_reduced, on='Date')

gasnat_reduced = gasnat[['Date', 'gasnat_last_5_avg', 'gasnat_last_3_avg', 'gasnat_last_1']]
merged_dfs = merged_dfs.merge(gasnat_reduced, on='Date')

gold_reduced = gold[['Date', 'gold_last_5_avg', 'gold_last_3_avg', 'gold_last_1']]
merged_dfs = merged_dfs.merge(gold_reduced, on='Date')

petrol_reduced = petrol[['Date', 'petrol_last_5_avg','petrol_last_3_avg', 'petrol_last_1']]
merged_dfs = merged_dfs.merge(petrol_reduced, on='Date')

rice_reduced = rice[['Date', 'rice_last_5_avg', 'rice_last_3_avg', 'rice_last_1']]
merged_dfs = merged_dfs.merge(rice_reduced, on='Date')

soy_reduced = soy[['Date', 'soy_last_5_avg', 'soy_last_3_avg', 'soy_last_1']]
merged_dfs = merged_dfs.merge(soy_reduced, on='Date')

wheat_reduced = wheat[['Date', 'wheat_last_5_avg', 'wheat_last_3_avg', 'wheat_last_1']]
merged_dfs = merged_dfs.merge(wheat_reduced, on='Date')


merged_dfs.head()
```

```python
correlation = merged_dfs.corr()

plt.figure(figsize=(20,12))
sns.heatmap(correlation, annot=True)
plt.show()
```
